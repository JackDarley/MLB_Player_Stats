```python
import pandas as pd

player_chart = pd.read_csv("player_ids.csv") 

position_dictionary = {}
number_of_ids = len(player_chart)
for i in range(number_of_ids):
    nb_of_columns = len(player_chart.values[i])
    for j in range(nb_of_columns):
        continue
    name = (player_chart.values[i][1])
    player_position = (player_chart.values[i][2])
    position_dictionary[name] = player_position

print(position_dictionary) 
#This code creates a dictionary to store a player's position
```

```python
# This program allows a user to input a position player's name to find out their career and 2020 season home run totals

from sportsreference.mlb.roster import Player # This imports all player stats from Sports Reference

import pandas as pd # This imports Pandas which will allow for better visual tables

player_homerun_dictionary = {} # This dictionary will store Home run totals in the 2020 season

while True:
    player_name = input("Enter an active player's name (enter DONE when done) ") 
    # Be sure to enter the name with correct spelling and capitilization!
    
    if player_name == "DONE":
        break
        # This means that you are done and do not wish to continue any further
        
    if player_name not in id_dictionary:
        print("This player did not play in the last three seasons or the name is not written correctly. Try a new player.")
        # The ID Dictionary includes players from the past 3 seasons only and requires names to be spelt with correct capitilization
        
    else:
        player_code = id_dictionary[player_name]
        player = Player(player_code)
        # These two lines look up a player's code and then store the code for future use
        
        
        if position_dictionary[player_name] == 'P':
            print(player_name,"is a pitcher.")
            # Since we are looking for position players only for home runs, we do not accept a pitcher here
            
        if position_dictionary[player_name] != 'P':
            print(player_name,"has",player.home_runs,"career home runs.")
            # This shows a player's career home run total
            
            player_homerun_dictionary[player_name] = player('2020').home_runs
            # This updates the Home Run dictionary
            
            if player('2020').games == 0:
                print("In 2020,",player_name,"hit 0 home runs.")
                # If a player did not play any games in 2020, they did not hit any home runs
                
            else:
                print("In 2020,",player_name,"hit",player('2020').home_runs, "home runs.")
                # This shows a player's 2020 home run total

hr_df = pd.DataFrame(list(player_homerun_dictionary.items()),columns = ['Player','2020 Home Runs'])
hr_df
# This creates the 2020 home run table
```

```python
# This program allows a user to input a pitcher's name to find out their career and 2020 season earned run average

from sportsreference.mlb.roster import Player # This imports all player stats from Sports Reference

import pandas as pd # This imports Pandas which will allow for better visual tables

player_era_dictionary = {} # This dictionary will store earned run averages in the 2020 season

while True:
    player_name = input("Enter an active player's name (enter DONE when done) ")
    # Be sure to enter the name with correct spelling and capitilization!
    
    if player_name == "DONE":
        break
        # This means that you are done and do not wish to continue any further
        
    if player_name not in id_dictionary:
        print("This player did not play in the last three seasons or the name is not written correctly. Try a new player.")
        # The ID Dictionary includes players from the past 3 seasons only and requires names to be spelt with correct capitilization
        
    else:
        player_code = id_dictionary[player_name]
        player = Player(player_code)
        # These two lines look up a player's code and then store the code for future use
        
        if position_dictionary[player_name] != 'P':
            print(player_name,"is not a pitcher.")
            # Since we are looking for pitchers only for earned run average, we do not accept a position player here
            
        if position_dictionary[player_name] == 'P':
            print(player_name,"has a",player.era,"career earned run average.")
            # This shows a pitcher's career earned run average
            
            player_era_dictionary[player_name] = player('2020').era
            # This updates the earned run average dictionary
            
            if player('2020').games == 0:
                print("In 2020,",player_name,"did not play in 2020.")
                # If a player did not play any games in 2020, they did not have an earned run average
            
            else:
                print("In 2020,",player_name,"had a",player('2020').era, "earned run average.")
                # This shows a pitcher's 2020 earned run average

era_df = pd.DataFrame(list(player_era_dictionary.items()),columns = ['Player','2020 Earned Run Average'])
era_df
# This creates the 2020 earned run average table
```

```python
#This program allows users to input a baseball player's from the past 3 seasons and shows position specific statistics from the 2020 season. 
#The comments below show which stats we record for each position:
#Outfielder: putouts
#First basemen: putouts
#Second basemen, third basemen, and short stop: assists
#Catcher: putouts
#Designated hitter: no stats
#Pitcher: assists

from sportsreference.mlb.roster import Player  # This imports all player stats from Sports Reference

while True:
    player_name = input("Enter an active player's name (enter DONE when done) ")
     # Be sure to enter the name with correct spelling and capitilization!
    
    if player_name == "DONE":
        break
         # This means that you are done and do not wish to continue any further
   
    if player_name not in id_dictionary:
        print("This player did not play in the last three seasons or the name is not written correctly. Try a new player.")
  # The ID Dictionary includes players from the past 3 seasons only and requires names to be spelt with correct capitilization  
    else:
        player_code = id_dictionary[player_name]
        player = Player(player_code)
         # These two lines look up a player's code and then store the code for future use
            
        if player('2020').games == 0:
            print("This player did not play in 2020")
        elif position_dictionary[player_name]== "CF":
            print(player_name, "is a center fielder and in 2020 he caught",player('2020').putouts, "fly balls")
        elif position_dictionary[player_name]== "RF":
            print(player_name, "is a right fielder and in 2020 he caught",player('2020').putouts, "fly balls")
        elif position_dictionary[player_name]== "LF":
            print(player_name, "is a left fielder and in 2020 he caught",player('2020').putouts, "fly balls")
        elif position_dictionary[player_name]== "OF":
            print(player_name, "is an outfielder and in 2020 he caught",player('2020').putouts, "fly balls")
         #These lines show how many putouts an outfielder had in 2020. 
        #If a player is a center fielder, right fielder, left fielder, or is listed as an outfielder in the database, he is an outfielder.      
            
        elif position_dictionary[player_name] == "1B":
            print(player_name, "is a first baseman and in 2020 he recorded", player('2020').putouts, "outs")
        #This shows how many outs a first baseman recorded in 2020
        
        elif position_dictionary[player_name] == "2B":
            print(player_name, "is a second baseman and in 2020 he threw out", player('2020').assists, "runners")
        elif position_dictionary[player_name] == "SS":
            print(player_name, "is a short stop and in 2020 he threw out", player('2020').assists, "runners")
        elif position_dictionary[player_name] == "3B":
            print(player_name, "is a third baseman and in 2020 he threw out", player('2020').assists, "runners")
         #These lines show many assists a second basemen, short stop and third basemen recorded   
            
        elif position_dictionary[player_name] == "C":
            print(player_name, "is a catcher and in 2020 he recorded", player('2020').putouts, "outs")
        #This shows how many putouts a catcher had in 2020
        
        elif position_dictionary[player_name] == "DH":
            print(player_name, "is a designated hitter and does not play defense")
        #Designated hitters do not play defense, so they have no defensive stats
        
        elif position_dictionary[player_name] == "P":
            print(player_name, "is a pitcher and in 2020 he threw out", player('2020').assists, "runners")
        #This shows how many assists a pitcher had in 2020. 
```
